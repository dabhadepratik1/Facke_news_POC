```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import nltk
import re
from gensim.parsing.preprocessing import remove_stopwords
import numpy as np
import pandas as pd
import joblib
```

```python
nltk.download('stopwords')
lemmatizer=WordNetLemmatizer()
stop_words=set(stopwords.words('english'))
```

```python
cv=joblib.load("cv.pkl")
ML_Model=joblib.load("RFC_Model.pkl")
```

```python
ML_Model
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
def text_cleaner(text):
  # Remove allother char except a-z and A-Z
  sent=re.sub('[^a-zA-z]',' ',text)

  # convert into lower case
  sent=sent.lower()

  # Lemmatization
  sent=" ".join([lemmatizer.lemmatize(word) for word in str(sent).split()])

  # remove Stopwords
  sent=" ".join([word for word in str(sent).split() if (word not in stop_words)])

  return sent
```

```python
#20800,
#"Specter of Trump Loosens Tongues, if Not Purse Strings, in Silicon Valley - The New York Times"
# 20800 0
# news is reliable
```

```python
raw_news="Specter of Trump Loosens Tongues, if Not Purse Strings, in Silicon Valley - The New York Times"
print("Before cleaning:- ",raw_news)

cleaned_news=text_cleaner(raw_news)
print("after cleaning :- ",cleaned_news)
```

```python
X=cv.transform([cleaned_news])
X
```

```python
pred=ML_Model.predict(X)[0]

if(pred==1):
  print("{} :- news is Fake".format(raw_news))

else:
  print("{} :- News is Real".format(raw_news))
```
